```python
import pandas as pd
import seaborn as sns
```

```python
articulatory_data = pd.read_csv('ACDS-volume.csv')
```

```python
articulatory_data
```

```python
articulatory_data['SVM'] = pd.to_numeric(articulatory_data['SVM'].str.strip('%'))
articulatory_data['PM'] = pd.to_numeric(articulatory_data['PM'].str.strip('%'))
articulatory_data
```

```python
x = sns.boxplot(
    x='Stage', y='Area', 
    data=articulatory_data, 
    showmeans=True, 
    order=["Presymptomatic", "Compensatory", "Post-Compensatory"])
```

```python
x = sns.boxplot(
    x='Stage', y='Volume', 
    data=articulatory_data, 
    showmeans=True, 
    order=["Presymptomatic", "Compensatory", "Post-Compensatory"])
```

```python
x = sns.boxplot(
    x='Stage', y='SVM', 
    data=articulatory_data, 
    showmeans=True, 
    order=["Presymptomatic", "Compensatory", "Post-Compensatory"])
```

```python
x = sns.boxplot(
    x='Stage', y='PM', 
    data=articulatory_data, 
    showmeans=True, 
    order=["Presymptomatic", "Compensatory", "Post-Compensatory"])
```

```python
g = sns.catplot(
    data=articulatory_data, kind="bar",
    x="Stage", y="Area",
    ci="sd", 
    palette="dark", 
    alpha=.6, height=6,
    order=["Presymptomatic", "Compensatory", "Post-Compensatory"]
)
```
